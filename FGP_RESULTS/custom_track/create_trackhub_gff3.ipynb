```python
# In[14]: Correct the short IDs field and extend the change to the uORFs ID field
def dividing(xy):
    '''
    this function is specific for joining the uORFs IDs
    '''
    y = xy[0]
    x = xy[1]
    splt_x = x.split(";")
    location = splt_x[1]
    ids = splt_x[0][9:]
    ids_splitted = ids.split(",")
    
    di = {}
    for ele in ids_splitted:
        di[splt_x[0][:9] + ele+";"+location] = y
    
    return di

xy_input = list(zip(uorfs_initial.index.values, uorfs_initial.uORF_id))
```

```python
dict_with_all = {}
size = 0
for i in xy_input:
#     print(i)
    dict_with_all.update(dividing(i))
    if len(dict_with_all) <= size:
        print("dup")
        break
    size = len(dict_with_all)
print(size)
```
